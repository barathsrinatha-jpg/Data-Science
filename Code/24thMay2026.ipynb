```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
x=[1,2,3,-4,-5,-6]
y=[10,20,30,40,50,60]
plt.scatter(x,y,c="#2ef098",s=200,alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("Number")
plt.ylabel("Values")
plt.show()
```

C = Colour ; S = Size ; Apha = Transparency between 0 to 1

```python
x=[1,2,3,4,5,6]
y=[10,27,36,44,53,60]
plt.figure(figsize=(10,5))
plt.plot(x,y,c="#f0552e",linestyle="--",linewidth="2")
plt.title("Line Plot")
plt.xlabel("Number")
plt.ylabel("Values")
plt.show()
```

Figure size should come before plot always

```python
x=['Men', 'Women', 'Children', 'Elders', 'Teens', 'Adults']
y=[10,27,36,44,53,60]
plt.figure(figsize=(10,5))
plt.bar(x,y,color="#f0552e")
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

Instead of C, use Color

```python
x=['Men', 'Women', 'Children', 'Elders', 'Teens', 'Adults']
y=[10,27,36,44,53,60]
plt.figure(figsize=(6,3))
colors = sns.color_palette("Set1")
plt.pie(y, labels=x, autopct="%1.3f%%", colors=colors)
plt.title("Pie Plot")
plt.show()
```

```python
data = sns.load_dataset("flights")
plt.figure(figsize=(10,5))
sns.lineplot(x="year", y="passengers", data=data,color="#f0552e")
plt.title("Line Plot")
plt.show()
data
```
