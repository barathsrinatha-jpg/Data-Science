```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.set_palette('rocket')
sns.boxplot(x='day',y='total_bill',data=df,hue='sex')
plt.show()
```

Hue is Like legend

```python
sns.set_palette('viridis')
sns.histplot(x='total_bill',data=df)
plt.show()
```

```python
data = sns.load_dataset('iris')
data
```

```python
plt.figure(figsize=(10, 5))
sns.set_palette('rocket')
sns.histplot(x='sepal_length', data=data, hue='species', kde=True, bins=20)
plt.show()
```

```python
sns.set_palette('viridis')
sns.violinplot(x='day',y='tip',data=df,hue='smoker')
plt.show()
```

```python
sns.set_palette('crest')
sns.countplot(x='species',data=data)
plt.show()
```

```python
sns.set_palette('crest')
sns.countplot(x='day',data=df,hue='smoker',stat='percent')
plt.show()
```

Stat shows whether you want Count or Percentage

```python
data1 = sns.load_dataset('penguins')
data1
```

```python
sns.set_palette('viridis')
sns.lmplot(x='bill_length_mm',y='bill_depth_mm',data=data1,hue='sex',markers=['o','x'])
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.set_palette('rocket')
sns.lmplot(x='bill_length_mm',y='bill_depth_mm',data=data1,hue='island',row='sex',col='species',markers=['o','x','s'])
plt.show()
```
