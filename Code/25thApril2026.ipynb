```python
# cannot be directly used in SQL
query = "insert into emp_details values(%s, %s, %s, %s)"
data = [(3, 'Harini', 90000, 28),
        (4, 'Aishwarya', 80000, 29),
    ]
cursor.executemany(query, data)
```
