```python
import statistics
```

```python
data = [80, 78, 82, 76, 81, 79, 77, 83, 80, 78]
```

```python
mean = statistics.mean(data)
```

```python
sd = statistics.stdev(data)
```

```python
population_sd = statistics.pstdev(data)
```

```python
print("Dataset:", data)
print("Mean:", mean)
print("Sample Standard Deviation:", sd)
print("Population Standard Deviation:", population_sd)
```

```python
!pip install scipy
```

```python
import scipy.stats as stats
```

```python
population_mean = 75
t_statistic, p_value = stats.ttest_1samp(data, population_mean)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
from scipy.stats import t
```

```python
df = 9
t_critical = t.ppf(1 - alpha/2, df)
print("T-Critical Value:", t_critical)
```

# 4th Test

```python
if p_value < Alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

T < Tcritical - Null Hypothesis is accepted

# Chi Square

```python
from scipy.stats import chisquare
```

```python
Observed = [40,30,20,30]
mean_observed = statistics.mean(Observed)
Expected = [mean_observed] * len(Observed)
print("Observed:", Observed)
print("Expected:", Expected)
```

If Expected is not given, then equal values

```python
chi2_statistic, p_value = chisquare(Observed, f_exp=Expected)
print("Chi-Square Statistic:", chi2_statistic)
print("P-value:", p_value)
```

```python
Alpha = 0.05
df_chi = len(Observed) - 1
chi2_critical = stats.chi2.ppf(1 - Alpha, df_chi)
print("Chi-Square Critical Value:", chi2_critical)
```

```python
if p_value < Alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```
