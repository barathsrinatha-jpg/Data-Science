```python
class Human:
    legs = 2
    arms = 2
    def __init__(self, name, age):
        self.name = name
        self.age = age
person= Human("Barath", 30)
print(f"Legs: {person.legs}, Arms: {person.arms}, Name: {person.name}, Age: {person.age}")
```

```python
class Human:
    def __init__(self, name, age,email,income):
        self.name = name
        self.age = age
        self.email = email
        self.income = income
    def get_tax(self,rate):
        self.tax = self.income * rate
    def is_elgible_for_license(self):
        if self.age >= 18:
            self.license = "Eligible for license"
        else:
            self.license = "Not eligible for license"
person= Human("Barath", 15, "barath@example.com", 50000)
print(person.__dict__)
person.get_tax(0.3)
print(person.__dict__)
person.is_elgible_for_license()
print(person.__dict__)
```
