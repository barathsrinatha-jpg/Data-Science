```python
df=df[df.columns[df.isna().sum()/len(df)<0.3]]
df.isnull().sum()
```

df=df[df.columns[df.isna().sum()/len(df)<0.3]] - Remember this

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.drop(["candidate_id","full_name"],axis=1,inplace=True)
df.head()
```

```python
df=pd.get_dummies(df,columns=["gender","college","degree","branch","platform","company_applied","sector","job_role","work_type","job_location"])
encoder=LabelEncoder()
df["gap_year"]=encoder.fit_transform(df["gap_year"])
df["prior_internship"]=encoder.fit_transform(df["prior_internship"])
df["referral_applied"]=encoder.fit_transform(df["referral_applied"])
df
```
