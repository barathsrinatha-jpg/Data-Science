```python
def operation():
    x = float(input("Enter the first number: "))
    y = float(input("Enter the second number: "))
    
    print("Select operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    
    choice = input("Enter choice (1/2/3/4): ")
    
    if choice == '1':
        return x + y
    elif choice == '2':
        return x - y
    elif choice == '3':
        return x * y
    elif choice == '4':
        if y != 0:
            return x / y
        else:
            return "Error: Division by zero"
    else:
        return "Invalid input"
operation()
```

```python
import json
n = int(input("Enter the number of employees: "))
employees = []
for i in range(n):
    name = input("Enter employee name: ")
    age = int(input("Enter employee age: "))
    city = input("Enter employee city: ")
    employees.append({"name": name, "age": age, "city": city})

with open(r"C:\Users\barat\Downloads\data.json", "w") as json_file:
    json.dump(employees, json_file,indent=4)
```

```python
with open(r"C:\Users\barat\Downloads\data.json", "r") as json_file:
    data = json.load(json_file)
    print(data)
```
