```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
df = pd.DataFrame()
df["Name"] = ["Alice", "Bob", "Charlie", "David"]
df["City"] = ["New York", "Los Angeles", "Chicago", "Houston"]
df["Gender"]=["Female","Male","Male","Female"]
df["Grades"]=["A","B","C","A"]
encoder = LabelEncoder()
df["Gender"] = encoder.fit_transform(df["Gender"])
df = pd.get_dummies(df, columns=["City"],dtype=int)
encoder = OrdinalEncoder(categories=[["C","B","A"]])
df["Encoded_Grades"] = encoder.fit_transform(df[["Grades"]]).ravel()
df
```

Outlier Removal Method - IQR

```python
Q1 = data["age"].quantile(0.25)
Q3 = data["age"].quantile(0.75)
IQR = Q3 - Q1
data["age"]=data["age"].clip(lower=Q1-1.5*IQR, upper=Q3+1.5*IQR)
sns.boxplot(x="age",data=data)
```

Clip the outliers in the "age" column using the IQR method and visualize the result with a boxplot.

IQR method is a technique used to identify and handle outliers in a dataset. It is based on the interquartile range (IQR), which is the difference between the third quartile (Q3) and the first quartile (Q1) of the data. The IQR method defines outliers as any data points that fall below Q1 - 1.5*IQR or above Q3 + 1.5*IQR. By using this method, we can effectively detect and manage outliers, which can help improve the accuracy of our analysis and models.
